#### Handling zero values
Checking for zero values in revenue, revenue_adj, budget, budget_adj and runtime as this might inconvenience the analysis if not handled. Zero values in these columns may indicate these information were not recorded for the movies.

```python
# Checking revenue, budget, runtime, revenue_adj, budget_adj column rows for zero values

zero_values= data.query('revenue==0 or budget == 0 or runtime ==0 or budget_adj ==0 or revenue_adj ==0').head()
zero_values.head()
```

```python
#replacing the zero values with NAN.

data['budget_adj'].replace(0, np.NAN, inplace=True)
data['budget'].replace(0, np.NAN, inplace=True)
data['revenue_adj'].replace(0, np.NAN, inplace=True)
data['revenue'].replace(0, np.NAN, inplace=True)
data['runtime'].replace(0, np.NAN, inplace=True)
```

#### Dropping missing values

```python
data.dropna(axis=0,inplace = True)
```

#### Changing release_date format to datetime format

```python
data['release_date'] = pd.to_datetime(data['release_date'])
```

#### Separating cast, director, genres, production_companies columns with pipe (|) into different columns

```python
#Creating new data frames to hold the separated columns

df_c = (data['cast'].str.split('|', expand= True).rename(columns= lambda x: f"cast_{x+1}"))
df_d = (data['director'].str.split('|', expand= True).rename(columns= lambda x: f"director_{x+1}"))
df_g = (data['genres'].str.split('|', expand= True).rename(columns= lambda x: f"genres{x+1}"))
df_p = (data['production_companies'].str.split('|', expand= True).rename(columns= lambda x: f"production_companies{x+1}"))

df_c.head(2)
df_p.head(2)
```

```python
# Joining the dataframes back to original dataset

data = data.join([df_c, df_d, df_g, df_p]) 
```

```python
# Dropping the original columns that were separated

data = data.drop(['cast', 'director', 'genres', 'production_companies'],axis = 1)
```

```python
# checking new df shape
data.shape
```

```python
# Checking the dataset
data.head()
```

```python
data.info()
```

```python
# Concatenating all the separated columns into a single column and dropping the NAN values

data['casts'] = data[data.columns[12:17]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)
data['directors'] = data[data.columns[17:29]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)
data['genres'] = data[data.columns[29:34]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)
data['production_companies'] = data[data.columns[34:39]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)

# Dropping the multiple columns

data.drop(data.iloc[:, 12:39],inplace= True,  axis = 1)
data.head(1)
```

The cleaned DataFrame has been duplicated into data2. All data types within the DataFrame are now correctly assigned. The dataset is prepared for exploratory data analysis. The cleaned dataset consists of 3,805 rows and 16 columns.

#### Question 1: In which year were the most movies released? How has the quantity of movies produced evolved over time?

```python
# Creating a line graph to visualize the change in the number of produced movies over the years

movies_year = data2.groupby('release_year').id.count()
ax = movies_year.plot.line(figsize=(10, 8), xticks=np.arange(1960, 2016, 5))

# Setting the graph labels
ax.set_title('Number of Released Movies by Year')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Released Movies')
value_2011 = movies_year.loc[2011]
ax.annotate(f"{int(value_2011)}", xy=(2011, value_2011), xytext=(2011, value_2011),
            fontsize=12, color='red', weight='bold')
```

#### Question 2: What are the most and least popular movies?

```python
def min_max_rows(data2, popularity):
    # Find the row of lowest popularity
    min_row = data2.loc[data2[popularity].idxmin()]
    
    # Find the row of highest popularity
    max_row = data2.loc[data2[popularity].idxmax()]
    
    # Concatenate the rows with the highest and lowest values and drop any NaN values
    result = pd.concat([max_row, min_row], axis=1).T.dropna()
    
    # Return the resulting DataFrame containing both rows
    return result

# Call the min_max_rows function with the DataFrame 'data2' and the column 'popularity'
result_df = min_max_rows(data2, 'popularity')

result_df
```

The most popular movie is "Jurassic World", releaased in 2015 and directed by Colin Trevorrow.

#### Question 3: In which month were the most movies released across all years?

```python
# Creating a new column called month from the release_date

data2['month'] = data2['release_date'].dt.month_name()
data2.head(1)
```

```python
# Group movies by month and count the number of movies for each month
movie_counts_by_month = data2.groupby('month')['id'].count()

# Create a new DataFrame to hold the months and the count of movies
month_movie_counts = pd.DataFrame({'month': movie_counts_by_month.index, 'count': movie_counts_by_month.values})

# Map month names to their corresponding numbers
month_number_map = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
                    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}
month_movie_counts['month_number'] = month_movie_counts['month'].map(month_number_map)

# Sort the DataFrame by the 'month_number' column
month_movie_counts = month_movie_counts.sort_values('month_number')

# Drop the 'month_number' column as it's no longer needed
month_movie_counts = month_movie_counts.drop('month_number', axis=1)

# Reset the index to get the desired numbering (0, 1, 2, 3, 4, ...)
month_movie_counts = month_movie_counts.reset_index(drop=True)

# Display the sorted DataFrame
month_movie_counts
```

Plotting a bar chart to show months and the number of movies

```python
# Set the plot style and size
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Plot a bar chart using the month_movie_counts DataFrame
ax = sns.barplot(x='month', y='count', data=month_movie_counts)

# Set the chart labels and title
ax.set(xlabel='', ylabel='Number of Movies', title='Number of Movies Released by Month')

# Get the highest month value (September) and its index
max_count = month_movie_counts['count'].max()
max_count_index = month_movie_counts['count'].idxmax()

# Add a data label for the highest month (September)
ax.annotate(f'{max_count}', xy=(max_count_index, max_count), xycoords='data',
            xytext=(0, 5), textcoords='offset points', ha='center', va='bottom',
            fontsize=10, color='black', weight='bold')

# Display the bar chart
plt.show()

```

In all the years, September holds the record for the highest number of movie releases, with a total of 452 movies produced.

#### Question 4: What are the 10 most and least profitable movies?

```python
# Adding a new column named profit
data2['profit']= data2['revenue'] - data2['budget']
data2.head(1)
```

```python
# Select the top 10 most profitable movies by sorting the 'profit' column in descending order
top_10_profit = data2[['title', 'profit']].sort_values('profit', ascending=False).head(10)

# Extract the titles and profits of the top 10 most profitable movies
x = top_10_profit['title'].tolist()
y = top_10_profit['profit'].tolist()

# Create a point plot for the top 10 most profitable movies
sns.set(rc={'figure.figsize': (8, 5)})
ax = sns.pointplot(x=y, y=x)

# Set the plot title and labels
ax.set_title("Top 10 Profitable Movies", fontsize=15)
ax.set_xlabel("Profit", fontsize=13)
plt.show()
```

```python
# Select the bottom 10 least profitable movies by sorting the 'profit' column in ascending order
bottom_10_profit = data2[['title', 'profit']].sort_values('profit', ascending=True).head(10)

# Extract the titles and profits of the 10 least profitable movies
x = bottom_10_profit['title'].tolist()
y = bottom_10_profit['profit'].tolist()

# Create a point plot for the 10 least profitable movies
sns.set(rc={'figure.figsize': (8, 5)})
ax = sns.pointplot(x=y, y=x)

# Set the plot title and labels
ax.set_title("Bottom 10 Profitable Movies", fontsize=15)
ax.set_xlabel("Profit", fontsize=13)
plt.show()
```

The highest-grossing movie is 'Avatar,' which amassed a total profit of $2,544,505,847. In contrast, 'The Warrior's Way' stands as the least profitable movie, with a total loss of $413,912,431.

#### Question 5: How are popularity, revenue, budget, and profit correlated with each other?

```python
# Creating a new DataFrame with only the relevant columns
corr_df = data2[['popularity', 'revenue', 'budget', 'profit']]

# Create a pairplot
sns.pairplot(corr_df)

# Set the title for the pairplot
plt.suptitle('Pairplot of Popularity, Revenue, Budget, and Profit', fontsize=15, y=1.02)

# Display the plot
plt.show()
```

The pairplot illustrates a positive correlation between revenue and profit, signifying that as a movie's revenue increases, its profit tends to rise as well.

The dataset itself posed limitations, as numerous entries had to be removed due to missing data. Additionally, certain columns like cast, genres, and directors required splitting and recombining due to the use of '|' as a separator, which increased the time taken for data analysis.
